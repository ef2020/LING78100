**Dynamic Typing**
Java: statically typed
* Variables are declared to refer to objects of a given type
* Methods use type signatures to enforce contracts

Python: dynamically typed
* Variables come into existence when first assigned to
* A variable can refer to an object of any type
* All types are (almost) treated the same way

**Assignment vs. “Is Equal to”**

In Math “=“ is used to say what is on the left equals what is on the right.
In Python, “=“ prescribes an action, “evaluate the expression on the right and assign its value to the variable named on the left.”

```
# why is the below set of commands invalid? 

r = 10
3.14*r*r = A
```

**The Key 2-step Action Behind Every Assignment Statement**

1. Evaluation the expression on the right hand side
2. Store the result in the variable named on the left hand side

* In algebra, t = t +10 doesn’t make sense unless you believe 0 = t-t = 10
* In Python, t = t + 10 means add 10 to the value of t and store the result in t.

```
t = 5
t = t+10
print (t)
```

**How do Python variables get values:**

* Assigned within the program (see the example above).
*  Users can submit values using the input() function.

**The input() function accepts an input from the user.**

When the input() function is called, the program flow will be stopped until the user has given an input and has ended the input with the return key.

The input() function reads a line entered on a console by an input device such as a keyboard, convert it into a string and returns it. For user-friendly formattin, finish the string with either white space of with the \n symbol. We will discuss the meaning of this when we get to strings.

Within the parenthesis of the input() function call you can specify a string that will be displayed on the computer screen before the program flow stops to wait for the input value from the user.

The input() function call should be on the RHS of the assignmen statement; the name of the variable whose value is expected from the used should be on the LHS of the assignment statement.

```
stName = input("What is your name?\n")  # request the user to submit her / his name
stBorough = input("In which NYC borough do your live? ")  # request the user to submit the value for the borough variable
print ("My name is", stName, ".") # print on the screen the current value of the name variable
print ("I live in", stBorough, ".") # print on the screen the current value of the borough variable
```
